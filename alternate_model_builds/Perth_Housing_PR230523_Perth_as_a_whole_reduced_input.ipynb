```python
import pandas as pd 
import numpy as np
house_sales = "Resources/Perth_Housing_Prices.csv"
sales_df = pd.read_csv(house_sales)
sales_df.set_index('ADDRESS', inplace=True)
sales_df.head()
```

```python
sales_df['DATE_SOLD'] = '01-' + sales_df['DATE_SOLD']
sales_df['DATE_SOLD'] = sales_df['DATE_SOLD'].str.replace('\r', '')
sales_df['DATE_SOLD'] = pd.to_datetime(sales_df['DATE_SOLD'])
sales_df.head()
```

```python
sales_df.dtypes
```

```python
sales_df['GARAGE'] = sales_df['GARAGE'].fillna(0)
```

```python
sales_df = sales_df.loc[sales_df['CBD_DIST'] <= 35000]
sales_df_a = sales_df.copy()
```

```python
# set y values
y = sales_df_a['PRICE']
```

```python
sales_df_a.columns
```

```python
sales_df_a = sales_df_a.drop(['SUBURB', 'PRICE', 'NEAREST_STN', 'DATE_SOLD', 'POSTCODE', 'LATITUDE', 'LONGITUDE', 'NEAREST_SCH', 'NEAREST_SCH_RANK', 'BUILD_YEAR', 'CBD_DIST', 'NEAREST_STN_DIST', 'NEAREST_SCH_DIST'], axis=1)
sales_df_a.head()
```

```python
columns = sales_df_a.columns
columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled = StandardScaler().fit_transform(sales_df_a)
scaled
```

```python
df_scaled = pd.DataFrame(scaled, columns=columns)
df_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_scaled, y)
```

```python
from sklearn.linear_model import LinearRegression
classifier = LinearRegression()
```

```python
 # Train the data
classifier.fit(X_train, y_train)
```

```python
prediction = classifier.predict(df_scaled)
```

```python
sales_df_b = sales_df.copy()
sales_df_b['Predicted Price'] = prediction
sales_df_b['Predicted Price'] = sales_df_b['Predicted Price'].round()
sales_df_b.head()
```

```python
print(f"Model's formula: y = {classifier.intercept_} + {classifier.coef_[0]}X")
```

```python
 # Import relevant metrics from scikit-learn: score, r2, mse, rmse, std
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Compute the metrics for the linear regression model
score = classifier.score(sales_df_a, y, sample_weight=None)
r2 = r2_score(y, prediction)
mse = mean_squared_error(y, prediction)
rmse = np.sqrt(mse)
std = np.std(y)

# Print releveant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```
