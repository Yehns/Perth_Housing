```python
import pandas as pd
```

```python
housing_data = pd.read_csv("all_perth_310121.csv")
housing_data
```

```python
sorted_df = housing_data.sort_values(by='BUILD_YEAR', ascending=False)

# Print the sorted DataFrame
sorted_df
```

```python
null_count = sorted_df["BUILD_YEAR"].isnull().sum()
null_count
```

```python
for i in range(len(sorted_df['BUILD_YEAR'])):
    if pd.isnull(sorted_df.loc[i, 'BUILD_YEAR']):
        sorted_df.loc[i, 'BUILD_YEAR'] = sorted_df.loc[i-1, 'BUILD_YEAR']
```

```python
for i in range(len(sorted_df['GARAGE'])):
    if pd.isnull(sorted_df.loc[i, 'GARAGE']):
        sorted_df.loc[i, 'GARAGE'] = 0
```

```python
sorted_df.reset_index(drop=True)
```
