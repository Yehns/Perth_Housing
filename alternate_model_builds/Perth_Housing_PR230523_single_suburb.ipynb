```python
import pandas as pd 
house_sales = "Resources/Perth_Housing_Prices.csv"
sales_df = pd.read_csv(house_sales)
sales_df.set_index('ADDRESS', inplace=True)
sales_df.head()
```

```python
sales_df.dtypes
```

```python
sales_df['GARAGE'] = sales_df['GARAGE'].fillna(0)
```

```python
check = sales_df.loc[sales_df['SUBURB'] == 'Roleystone']
check.head()
```

```python
sales_df_a = sales_df.copy()
check_2 = check.copy()
```

```python
sales_df_a.columns
```

```python
sales_df_a = sales_df_a.drop(['NEAREST_STN', 'DATE_SOLD', 'POSTCODE', 'LATITUDE', 'LONGITUDE', 'NEAREST_SCH', 'NEAREST_SCH_RANK'], axis=1)
sales_df_a.head()
```

```python
check_2 = check_2.drop(['SUBURB', 'PRICE', 'NEAREST_STN', 'DATE_SOLD', 'POSTCODE', 'LATITUDE', 'LONGITUDE', 'NEAREST_SCH', 'NEAREST_SCH_RANK', 'BUILD_YEAR', 'CBD_DIST', 'NEAREST_STN_DIST', 'NEAREST_SCH_DIST'], axis=1)
check_2.head()
```

```python
# set y values
#y = sales_df_a['PRICE']
y = check['PRICE']
```

```python
columns = check_2.columns
columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
check_2.head()
```

```python
scaled = StandardScaler().fit_transform(check_2)
scaled
```

```python
df_scaled = pd.DataFrame(scaled, columns=columns)
df_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_scaled, y)
```

```python
from sklearn.linear_model import LinearRegression
classifier = LinearRegression()
```

```python
 # Train the data
classifier.fit(X_train, y_train)
```

```python
prediction = classifier.predict(df_scaled)
```

```python
check_3 = check.copy()
check_3['Predicted Price'] = prediction
check_3['Predicted Price'] = check_3['Predicted Price'].round()
check_3.head()
```

```python
print(f"Model's formula: y = {classifier.intercept_} + {classifier.coef_[0]}X")
```

```python
 # Create a line plot of the predicted salary values
import hvplot.pandas
best_fit_line = check_3.hvplot.scatter(
    x = "PRICE",
    y = "Predicted Price",
    color = "red"
)
best_fit_line
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a, b = np.polyfit(check_3['PRICE'], check_3['Predicted Price'], 1)

plt.figure(figsize=(15,8))
plt.scatter(check_3['PRICE'], check_3['Predicted Price'], marker="o", facecolors="red", edgecolors="black")
plt.plot(check_3['PRICE'], a*check_3['PRICE']+b)
plt.title('Roleystone')
plt.ticklabel_format(style='plain')
plt.show()
```

```python
 # Import relevant metrics from scikit-learn: score, r2, mse, rmse, std
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Compute the metrics for the linear regression model
score = classifier.score(check_2, y, sample_weight=None)
r2 = r2_score(y, prediction)
mse = mean_squared_error(y, prediction)
rmse = np.sqrt(mse)
std = np.std(y)

# Print releveant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is {std}.")
```
