```python
import pandas as pd 
import io 
from google.colab import files 
uploaded = files.upload()
df = pd.read_csv(io.BytesIO(uploaded['all_perth_310121.csv']))
df.head()
```

```python
df.dtypes
```

```python
df['GARAGE'] = df['GARAGE'].fillna(0)
```

```python
import numpy as np
df['BUILD_YEAR'] = df['BUILD_YEAR'].fillna(df['BUILD_YEAR'].mean())
#df['BUILD_YEAR'] = pd.to_numeric(df['BUILD_YEAR'])
df['BUILD_YEAR'] = df['BUILD_YEAR'].round()
```

```python
check = df.loc[df['SUBURB'] == 'Roleystone']
check
```

```python
df.dtypes
```

```python
df_a = df.copy()
```

```python
df_a.columns
```

```python
df_a = df_a.drop(['ADDRESS', 'NEAREST_STN', 'DATE_SOLD', 'POSTCODE', 'LATITUDE', 'LONGITUDE', 'NEAREST_SCH', 'NEAREST_SCH_RANK'], axis=1)
df_a.head()
```

```python
# set y values
y = df_a['PRICE']
```

```python
dummies = pd.get_dummies(df_a['SUBURB'])
dummies.head()
```

```python
df_merged = pd.concat([df_a, dummies], axis=1)
df_merged.head()
```

```python
df_merged = df_merged.drop(['SUBURB', 'PRICE'], axis=1)
df_merged.head()
```

```python
columns = df_merged.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled = StandardScaler().fit_transform(df_merged)
scaled
```

```python
df_scaled = pd.DataFrame(scaled, columns=columns)
df_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_scaled, y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs')
```

```python
 # Train the data
classifier.fit(X_train, y_train)
```
